```python
#Compare two numbers
num1 = float(input("Type the first number: "))
num2 = float(input("Type the second number: "))
if num1>num2:
    print(f"{num1} > {num2}")
elif num1==num2:
    print(f"{num1} = {num2}")
else:
    print(f"{num2} > {num1}")
```

```python
#Second grade equation
a = float(input("Type the coefficient a:"))
b = float(input("Type the coefficient b:"))
c = float(input("Type the coefficient c:"))
delta = b**2 - 4*a*c
x1 = (-b + delta**(0.5))/(2*a)
x2 = (-b - delta**(0.5))/(2*a)
if delta <0:
    print(f"Delta is less than zero:{delta}")
    print("There's no real root.")
elif a == 0:
    print(f"{a} is less than zero, insert a number different than zero")
elif delta >= 0:
    print(f"x' = {x1} = x'' = {x2}")
else:
    print(f"x' = {x1} e x'' = {x2}")
    
```
