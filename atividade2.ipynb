```python
10. Sistema simples de login
Peça ao usuário nome e senha. Considere:
▶ usuario_correto = "admin"
▶ senha_correta = "1234"
Se os dados estiverem corretos, mostre "Login
realizado com sucesso". Caso contrário, "Usuário ou
senha incorretos". (Use o operador ==).
```

```python
usuario_correto = "admin"
senha_correta = "1234"
usuario = input("Digite seu usuário: ")
senha = input("Digite sua senha: ")
if usuario == usuario_correto and senha == senha_correta:
    print("Login efetuado com sucesso.")
else:
    print("Usuário ou senha incorretos.")
```

```python
Crie um programa que peça um número e mostre:
▶ Número positivo
▶ Número negativo
▶ Número igual a zero
Dica: Use as estruturas if, elif e else para resolver
esse problema.
```

```python
numero = int(input("Digite seu número: "))
if numero > 0:
    print("Número positivo")
elif numero < 0:
    print("Número negativo")
else:
    print("Número igual a zero")
```
