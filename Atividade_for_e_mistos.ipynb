```python
pares = 0
for x in range(5):
    num = int(input(f"Digite o {x+1}º número: "))
    if num % 2 == 0:
        pares += 1

print(f"Você digitou {pares} números pares.")
```

```python
nota = -1
while nota < 0 or nota > 10:
    nota = float(input("Digite uma nota entre 0 e 10: "))
    if nota < 0 or nota > 10:
        print("Valor inválido! Tente novamente.")

print(f"Nota {nota} registrada com sucesso.")
```

```python
usuario_correto = "admin"
senha_correta = "1234"
tentativas = 3

while tentativas > 0:
    user = input("Usuário: ")
    senha = input("Senha: ")

    if user == usuario_correto and senha == senha_correta:
        print("Login realizado com sucesso! Bem-vindo.")
        break
    else:
        tentativas -= 1
        if tentativas > 0:
            print(f"Erro! Você ainda tem {tentativas} tentativa(s).")
        else:
            print("Conta bloqueada após 3 tentativas falhas.")
```
